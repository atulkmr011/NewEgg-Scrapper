```python
from bs4 import BeautifulSoup # Importing the BeautifulSoup package
import requests # Importing requests library
import csv
import pandas as pd
from time import sleep
from random import randint
```

```python
#here, enter the number of ages to be scrapped
lap_nm=[]
lap_pr=[]
page_num=input("Enter Number of page")
for i in range (1,int(page_num)+1):
    url=("https://www.newegg.com/p/pl?d=laptop+refurbish&N=50001759%2050010772%2050001186%2050001009%2050001315%2050010418%2050001149%2050001146%2050001312%2050001077%2050001259&PageSize=96&page=")+str(i) # Requesting the content of the URL
    req = requests.get(url)
    content=BeautifulSoup(req.content,'html.parser')
    name=content.find_all('a' ,{'class':'item-title'})
    price = content.find_all('li',{'class':'price-current'})
    sleep(randint(2,8))
    print("Laptop in page"+str(i))
    print(len(name))
    
    for i in name:
        lap_nm.append(i.text)
    for i in price:
        lap_pr.append(i.text)
```

```python
for i in lap_pr:
    print(i)
```

```python
#number of laptop name
len(lap_nm)
```

```python
#number of laptop price
len(lap_pr)
```

```python
data={"description":lap_nm,"price":lap_pr}
df=pd.DataFrame(data)
print(df)
```
